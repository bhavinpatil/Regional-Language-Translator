```python
import xmltodict
import pandas as pd
```

```python
with open('dataset.tmx', 'r', encoding='utf-8') as tmx_file:
    tmx_data = tmx_file.read()
```

```python
tmx_dict = xmltodict.parse(tmx_data)
```

```python
translation_pairs = []
for tu in tmx_dict['tmx']['body']['tu']:
    source_text = tu['tuv'][0]['seg']
    target_text = tu['tuv'][1]['seg']
    translation_pairs.append((source_text, target_text))
```

```python
df = pd.DataFrame(translation_pairs, columns=['Hindi', 'Telugu'])
```

```python
df.to_csv('processed_data.csv', index=False, encoding='utf-8')
```
