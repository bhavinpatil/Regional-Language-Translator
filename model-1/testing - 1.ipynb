```python
import pickle
from tensorflow.keras.models import load_model

# Load the trained model
loaded_model = load_model('models/translation_model3.keras')

# Load the Hindi tokenizer
with open('tokenizers/hindi_tokenizer3.pkl', 'rb') as tokenizer_file:
    tokenizer_hindi = pickle.load(tokenizer_file)

# Load the Telugu tokenizer
with open('tokenizers/telugu_tokenizer3.pkl', 'rb') as tokenizer_file:
    tokenizer_telugu = pickle.load(tokenizer_file)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# sample_input_sentence1 = "मैं आपको पसंद करता हूँ"
sample_input_sentence1 = "यह विचार बहुत ही सरल है"
sample_input_sentence2 = "ఈ ఆలోచన చాలా సులభమైనది"
```

```python
sample_input_tokens = tokenizer_hindi.texts_to_sequences([sample_input_sentence1])
sample_output_tokens = tokenizer_telugu.texts_to_sequences([sample_input_sentence2])
```

```python
sample_input_tokens, sample_output_tokens
```

```python
# Pad the input sequence to match the model's input shape
sample_input_padded = pad_sequences(sample_input_tokens, maxlen=90, padding='post')
sample_input_padded
```

```python
decoder_input_data = np.zeros((1, 90))
decoder_input_data
```

```python
sample_output_sentence = []

# Set the initial decoder input token to the start token
decoder_input_token = tokenizer_telugu.word_index['<OOV>']

# Define a maximum sequence length to avoid infinite loops
max_sequence_length = 90

for _ in range(max_sequence_length):
    # Predict the next word in the output sequence
    predicted_word_index = loaded_model.predict([sample_input_padded, np.array([[decoder_input_token]])]).argmax(axis=-1)[0, 0]
    
    # Map the predicted word index to the corresponding word
    predicted_word = tokenizer_telugu.index_word.get(predicted_word_index, '<OOV>')
    
    # If the predicted word is the end token, stop generating
    if predicted_word == '<OOV>':
        break
    
    # Append the predicted word to the output sequence
    sample_output_sentence.append(predicted_word)
    
    # Update the decoder input with the predicted word index for the next iteration
    decoder_input_token = predicted_word_index

# Join the predicted words to form the output sentence
generated_telugu_sentence = ' '.join(sample_output_sentence)

# Print the generated Telugu output sentence
print("Generated Telugu Sentence:", generated_telugu_sentence)
```
